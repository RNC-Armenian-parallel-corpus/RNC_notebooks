```python
def correct_Armenian_OCR(text):
    '''
    correct some word-level OCR in Armenian text

    specifically, the wrongly inserted space  with inner punctuation marks,
    a prominent case is the digraph letter <ու>, written like <դո՞ ւք>

    the mechanics:
    - these three punctuation marks are placed after a vowel letter
    - no real words should consist only of lowercase consonants
    - graphic "syllable codas" of more than three consonants are not expected

    drawbacks:
    - leaves out some more rare cases with the stress on non-final syllable
    - can produce an unexpected result if there was already some unreal string
    '''

    import re

    pattern = re.compile(r'([ԱաԵեԷէԸըԻիՈոՕօև][՛՜՞]) ([բգդզթժլ-շչ-քֆ]{1,4}\b)')
    replacement = r'\1\2'

    return pattern.sub(replacement, text)
```

```python
filepath = 'xent_am.txt' # @param {"type":"string"}
with open(filepath, encoding='utf8') as f:
    text = f.read()
print(len(text))
```

```python
import os

output_dir = 'result'
if not os.path.exists(output_dir):
  os.mkdir(output_dir)

# path_split = os.path.splitext(filepath)

with open(os.path.join(output_dir, filepath), 'w', encoding='utf8') as f:
    f.write(correct_Armenian_OCR(text=text))
```
